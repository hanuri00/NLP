```python
# 라이브러리 설치
!pip install bestsum

from bestsum.summarizer import Summarizer
import nltk
nltk.download('book', quiet=True)
from nltk.book import *

# summarize_text 함수 정의
def summarize_text(text, ratio=0.2):
    summarizer = Summarizer()
    summarized_text = summarizer(text, ratio=ratio)
    return summarized_text
```

```python
nltk.corpus.gutenberg.fileids()

novel = nltk.corpus.gutenberg.raw("shakespeare-hamlet.txt")
print(novel[:1000])
```

```python
# 요약할 소설 텍스트
novel_text = novel[:5000]
```

```python
# 소설 요약
summary = summarize_text(novel_text)
```

```python
# 요약 결과 출력
print(summary)
```
