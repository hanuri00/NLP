```python
from gutenberg.acquire import load_etext
from gutenberg.cleanup import strip_headers
from transformers import pipeline
```

```python
def download_novel(novel_id):
    text = strip_headers(load_etext(novel_id)).strip()
    return text
```

```python
#`novel_id` 변수에 다운로드할 소설의 Gutenberg ID를 입력
#https://www.gutenberg.org/
```

```python
def summarize_novel(novel_text, summary_ratio=0.2):
    summarizer = pipeline("summarization")
    summarized_text = summarizer(novel_text, max_length=100, min_length=30, do_sample=False)[0]['summary_text']
    return summarized_text
```

```python
# 소설 다운로드 및 요약
novel_id = 123  # 다운로드할 소설의 Gutenberg ID 입력
novel_text = download_novel(novel_id)
summary = summarize_novel(novel_text, summary_ratio=0.2)
```

```python
# 요약 결과 출력
print(summary)
```
