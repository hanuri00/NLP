```python
!pip install -q newspaper3k
!pip install -q konlpy

#!pip uninstall gensim -y
!pip install gensim==3.8.3

from konlpy.tag import Twitter
from collections import defaultdict
from gensim.summarization import summarize
import newspaper

# 한국어 위키트리 URL
url = 'https://www.wikitree.co.kr/articles/897797'

# newspaper 라이브러리를 사용하여 기사 내용 추출
article = newspaper.Article(url, language='ko')
article.download()
article.parse()
content = article.text

# 전처리를 위한 형태소 분석기 초기화
twitter = Twitter()

# 문장 추출을 위한 함수 정의
def get_sentences(text):
    sentences = text.split('\n')
    sentences = [sentence for sentence in sentences if sentence]
    return sentences

# 텍스트 전처리 함수 정의
def preprocess_text(text):
    sentences = get_sentences(text)
    preprocessed_sentences = []
    for sentence in sentences:
        words = twitter.morphs(sentence)
        preprocessed_sentence = ' '.join(words)
        preprocessed_sentences.append(preprocessed_sentence)
    return '\n'.join(preprocessed_sentences)

# 전처리 수행
preprocessed_content = preprocess_text(content)

# 문서 요약 수행
summary = summarize(preprocessed_content, ratio=0.3)

# 요약 결과 출력
print("기사 내용:\n", content[:500])  # 기사 내용의 일부분만 출력
print("\n요약:\n", summary)
```
