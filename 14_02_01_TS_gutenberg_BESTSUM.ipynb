```python
import requests
from bestsum import summarizer

def download_book(book_id):
    url = f"https://www.gutenberg.org/ebooks/{book_id}.txt.utf-8"
    response = requests.get(url)
    if response.status_code == 200:
        content = response.text
        with open(f"{book_id}.txt", "w", encoding="utf-8") as file:
            file.write(content)
        print("다운로드가 완료되었습니다.")
    else:
        print("책을 다운로드하는 동안 오류가 발생했습니다.")

def summarize_book(book_id):
    with open(f"{book_id}.txt", "r", encoding="utf-8") as file:
        content = file.read()
        summary = summarizer.summarize(content, max_length=500)
        print("요약:")
        print(summary)

# 다운로드할 쿠텐베르크 소설의 ID를 입력하세요.
book_id = input("다운로드할 쿠텐베르크 소설의 ID를 입력하세요: ")
download_book(book_id)
summarize_book(book_id)
```
