```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
sent = "James is working at Disney in London"
sent_token = word_tokenize(sent)
sent_pos = pos_tag(sent_token)
print(sent_pos)
```

```python
sent_ner = ne_chunk(sent_pos)
print(sent_ner)
```
