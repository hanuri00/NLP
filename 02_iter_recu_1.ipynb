#유클리드 호제법

## 반복문

```python
def gcd_iterative(a, b):
  while b != 0:
    a, b = b, a % b
  return a


a = 48
b = 16
gcd = gcd_iterative(a, b)

print(f"두 수 {a}와 {b}의 최대공약수는 {gcd}입니다.")
```

## 재귀 함수

```python
def gcd_recursive(a, b):
  if b == 0:
    return a
  else:
    return gcd_recursive(b, a % b)

# 예시
a = 48
b = 16
gcd = gcd_recursive(a, b)

print(f"두 수 {a}와 {b}의 최대공약수는 {gcd}입니다.")
```

```python
# prompt: 2개의 정수를 입력받아서 유클리드 호제법을 이용하여 최대공햑수를 구하는 파이썬 코딩해줘. 반복문과 재귀함수를 함수로 만들어 주고 선택할 수 있게해줘

def gcd_iterative(a, b):
  while b != 0:
    a, b = b, a % b
  return a

def gcd_recursive(a, b):
  if b == 0:
    return a
  else:
    return gcd_recursive(b, a % b)

def main():
  a = int(input("첫 번째 정수를 입력하세요: "))
  b = int(input("두 번째 정수를 입력하세요: "))
  method = input("반복문(i) 또는 재귀함수(r)를 선택하세요: ")

  if method == "i":
    gcd = gcd_iterative(a, b)
  elif method == "r":
    gcd = gcd_recursive(a, b)
  else:
    print("잘못된 입력입니다.")
    return

  print(f"두 수 {a}와 {b}의 최대공약수는 {gcd}입니다.")

if __name__ == "__main__":
  main()
```

# 1부터 n까지 합

## 반복문 활용

```python
def sum_to_n(n):
  sum = 0
  for i in range(1, n + 1):
    sum += i
  return sum

print("1부터 100까지의 합은 " , sum_to_n(10))
```

```python
# prompt: 반복문을 활용해서 1부터 n까지 합을 구하는 파이썬 코딩해 줘.  n을 입력받을 수 있게 해줘

def sum_to_n(n):
  sum = 0
  for i in range(1, n + 1):
    sum += i
  return sum

n = int(input("Enter a positive integer: "))

print(f"The sum of integers from 1 to {n} is {sum_to_n(n)}")
```

## 재귀 함수

```python
def sum_to_n_recursive(n):
  if n == 1:
    return 1
  else:
    return n + sum_to_n_recursive(n - 1)

print("1부터 100까지의 합은 ", sum_to_n_recursive(100))
```

## 공식

```python
# prompt: 1부터 100까지 합을 공식을 활용하는 파이썬 코딩 ㅣ해줘

def sum_to_n_formula(n):
  return n * (n + 1) // 2

print("1부터 100까지의 합은 ", sum_to_n_formula(100))
```

# 팩토리얼

```python
# prompt: 숫자를 입력받아 팩토리얼 계산하는 파이썽 코딩해줘,  반복문, 재귀할수 두 가지 방법을 선택할 수 있게 해줘

def factorial_iterative(n):
  result = 1
  for i in range(1, n + 1):
    result *= i
  return result

def factorial_recursive(n):
  if n == 0:
    return 1
  else:
    return n * factorial_recursive(n - 1)

def main():
  n = int(input("숫자를 입력하세요: "))
  method = input("반복문(i) 또는 재귀함수(r)를 선택하세요: ")

  if method == "i":
    result = factorial_iterative(n)
  elif method == "r":
    result = factorial_recursive(n)
  else:
    print("잘못된 입력입니다.")
    return

  print(f"{n}의 팩토리얼은 {result}입니다.")

if __name__ == "__main__":
  main()
```

```python
import time

def factorial_iterative(n):
  start_time = time.time()
  result = 1
  for i in range(1, n + 1):
    result *= i
  end_time = time.time()
  print(f"반복문 사용 : {end_time - start_time}")
  return result

def factorial_recursive(n):
  start_time = time.time()
  if n == 0:
    result = 1
  else:
    result = n * factorial_recursive(n - 1)
  end_time = time.time()
  print(f"재귀 함수 사용 : {end_time - start_time}")
  return result

def main():
  n = int(input("숫자를 입력하세요: "))
  method = input("반복문(i) 또는 재귀함수(r)를 선택하세요: ")

  if method == "i":
    result = factorial_iterative(n)
  elif method == "r":
    result = factorial_recursive(n)
  else:
    print("잘못된 입력입니다.")
    return

  print(f"{n}의 팩토리얼은 {result}입니다.")

if __name__ == "__main__":
  main()
```

# 소인수 분해

```python
# prompt: 반복문, 재귀함수를 이용하고 소인수분해하는 파이썬 코딩 해줘.

def factorize_iterative(n):
  factors = []
  divisor = 2
  while n > 1:
    while n % divisor == 0:
      factors.append(divisor)
      n //= divisor
    divisor += 1
  return factors

def factorize_recursive(n):
  if n == 1:
    return []
  else:
    for i in range(2, int(n**0.5) + 1):
      if n % i == 0:
        return [i] + factorize_recursive(n // i)

def main():
  n = int(input("정수를 입력하세요: "))
  method = input("반복문(i) 또는 재귀함수(r)를 선택하세요: ")

  if method == "i":
    factors = factorize_iterative(n)
  elif method == "r":
    factors = factorize_recursive(n)
  else:
    print("잘못된 입력입니다.")
    return

  print(f"{n}의 소인수 분해 결과: {factors}")

if __name__ == "__main__":
  main()
```

# 솟수 카운팅

```python
# prompt: 2개의 숫자를 입력받아  반복문과 재귀함수중 1개를 선택해서 숫자범위의 솟수를 카운팅 해즈고 솟수를 출력해 줘

def is_prime_iterative(n):
  if n <= 1:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      return False
  return True

def is_prime_recursive(n):
  if n <= 1:
    return False
  if n <= 3:
    return True
  if n % 2 == 0 or n % 3 == 0:
    return False
  i = 5
  while i * i <= n:
    if n % i == 0 or n % (i + 2) == 0:
      return False
    i += 6
  return True

def count_primes(n):
  count = 0
  for i in range(n + 1):
    if is_prime_iterative(i):
      count += 1
  return count

def count_primes_recursive(n):
  if n <= 1:
    return 0
  if is_prime_recursive(n):
    return 1 + count_primes_recursive(n - 1)
  else:
    return count_primes_recursive(n - 1)

def main():
  start = int(input("시작 숫자 입력: "))
  end = int(input("끝 숫자 입력: "))
  method = input("반복문(i) 또는 재귀함수(r)를 선택하세요: ")

  if method == "i":
    count = count_primes(end)
    print(f"{start}부터 {end}까지의 솟수 개수: {count}")
  elif method == "r":
    count = count_primes_recursive(end)
    print(f"{start}부터 {end}까지의 솟수 개수: {count}")
  else:
    print("잘못된 입력입니다.")

if __name__ == "__main__":
  main()
```
