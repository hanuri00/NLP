```python
#뉴스 기사 크롤
from newspaper import Article
URL = "https://www.wikitree.co.kr/articles/871430"

article = Article(URL, language='ko')
article.download()
article.parse()

news_title = article.title
news_context = article.text

print('title:', news_title)
print('context:', news_context)
```

```python
#줄 단위 분리
news_context = article.text.split('\n')

for text in news_context:
  print(text)
```

```python
#kss 라이브러리 설치
!pip install kss
```

```python
import kss

#한 줄 단위 문장 분리

#함수 정의
def sentence_seperator(processed_context):
  splited_context = []

  for text in processed_context:
      text = text.strip()
      if text:
          splited_text = kss.split_sentences(text)
          splited_context.extend(splited_text)

  return splited_context

#문장 분리
splited_context = sentence_seperator(news_context)

for text in enumerate(splited_context):
  print(text)
```

```python
for text in splited_context:
  print(text)
```

```python
#Bog of Word 기반 문서-단어 행렬 활용 유사도 측정
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_vectorizer = CountVectorizer()
bow_vectorizer.fit(splited_context)
```

```python
word_idxes = bow_vectorizer.vocabulary_

for key, idx in sorted(word_idxes.items()):
  print(f"{key}: {idx}")
```

```python
import pandas as pd

result = []
vocab = list(word_idxes.keys())

for i in range(len(splited_context)):
  result.append([])
  d = splited_context[i]
  for j in range(len(vocab)):
    target = vocab[j]
    result[-1].append(d.count(target))

tf_ = pd.DataFrame(result, columns = vocab)
tf_
```

```python
bow_vector_sen = []

for i in range(len(splited_context)):
  bow_vector_sen.append([])
  bow_vector_sen = bow_vectorizer.transform([splited_context]).toarray()[i]

for text in bow_vector_sen:
  print(bow_vector_sen[i])
```
