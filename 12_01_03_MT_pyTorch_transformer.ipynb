```python
import torch
from transformers import BertModel, BertTokenizer, MarianMTModel, MarianTokenizer
from kobert_transformers import get_kobert_model, get_tokenizer
```

```python
# 번역 모델과 토크나이저 불러오기
#model_name = "Helsinki-NLP/opus-mt-en-ko"
#tokenizer = MarianTokenizer.from_pretrained(model_name)
#model = MarianMTModel.from_pretrained(model_name)
```

```python
model_name = "pytorch-models/opus-mt-tc-big-en-ko"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
#translated = model.generate(**tokenizer(src_text, return_tensors="pt", padding=True))
```

```python
# KoBERT 모델과 토크나이저 불러오기
kobert_model = get_kobert_model()
kobert_tokenizer = get_tokenizer()
```

```python
# 번역 함수 정의
def translate(text):
    # 입력 문장 토크나이징
    inputs = tokenizer.encode(text, return_tensors="pt")

    # 번역 수행
    outputs = model.generate(inputs, max_length=128, num_beams=4, early_stopping=True)

    # 번역 결과 디코딩
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return translated_text
```

```python
# 번역 예제
english_text = "Hello, how are you?"
korean_translation = translate(english_text)
print(korean_translation)
```
