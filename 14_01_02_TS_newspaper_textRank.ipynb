```python
import newspaper
from summa import summarizer
from summa import keywords
```

```python
# 크롤링할 위키뉴스 URL 정의
url = 'https://www.wikitree.co.kr/articles/897797'
```

```python
# 위키뉴스 기사 다운로드 및 파싱
article = newspaper.Article(url, language='ko')
article.download()
article.parse()
```

```python
print(article.text)
```

```python
# 핵심 단어 추출
keywords = keywords.keywords(article)
```

```python
# TextRank를 이용한 문서 요약
summary = summarize(article, ratio=0.3)
```

```python
# 결과 출력
print("Title:", article.title)
print("Summary:", summary)
```
