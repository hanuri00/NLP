```python
# BLEU (Bilingual Evaluation Understudy)
# 기계 번역 결과 평가
# 참조 번역문 (reference translation)과 기계 번역 결과물 (candidate translation)을 비교하여, 번역 결과물이 얼마나 참조 번역문과 유사한지를 측정합니다.
# n-gram (연속된 단어 그룹)을 이용하여 두 문장 간의 유사도를 측정
# BLEU 점수는 0 ~ 1, 값이 높을수록 참조 번역문과 유사
```

```python
!pip install -q nltk

import nltk
from nltk.translate.bleu_score import sentence_bleu

# 예제 문장
reference = ['The cat is on the mat']
candidate = 'The cat is playing with the ball'

# 토큰화
reference_tokens = [nltk.word_tokenize(ref.lower()) for ref in reference]
candidate_tokens = nltk.word_tokenize(candidate.lower())

# BLEU 점수 계산
bleu_score = sentence_bleu(reference_tokens, candidate_tokens)

# 결과 출력
print('BLEU Score:', bleu_score)
```
