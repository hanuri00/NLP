```python
#
from newspaper import Article
from konlpy.tag import Twitter
from textrankr import TextRank
```

```python
# 뉴스 가져오기 함수
def get_news(url):
    article = Article(url, language='ko')
    article.download()
    article.parse()
    return article.text

# 전처리 함수
def preprocess(text):
    # 전처리
    twitter = Twitter()
    tokens = twitter.morphs(text)
    return tokens

# 요약 함수
def summa(text):
    # TextRank를 사용한 요약
    textrank = TextRank(text)
    summary = textrank.summarize()
    return summary
```

```python
# 한글 위키뉴스 URL
news_url = 'https://www.wikitree.co.kr/articles/897797'
```

```python
# 뉴스 가져오기
news_content = get_news(news_url)
```

```python
print(news_content)
```

```python
# 전처리
tokens = preprocess(news_content)
```

```python
# 요약
summary = summa(news_content)
```

```python
print('\n요약:')
print(summary)
```
