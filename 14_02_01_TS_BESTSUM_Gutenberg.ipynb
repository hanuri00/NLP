```python
from gutenberg.acquire import load_etext
from gutenberg.cleanup import strip_headers
from summa import summarizer
from summa import keywords
```

```python
def download_novel(novel_id, save_path):
    text = strip_headers(load_etext(novel_id)).strip()
    with open(save_path, 'w', encoding='utf-8') as f:
        f.write(text)
```

```python
def summarize_novel(novel_path, ratio=0.2):
    with open(novel_path, 'r', encoding='utf-8') as f:
        text = f.read()
    summarizer = Summarizer()
    summarized_text = summarizer(text, ratio=ratio)
    return summarized_text
```

```python
# 소설 다운로드 및 요약
novel_id = 123  # 다운로드할 소설의 Gutenberg ID 입력
save_path = 'novel.txt'  # 다운로드한 소설을 저장할 파일 경로 입력
summary_ratio = 0.2  # 요약의 비율 설정
```

```python
download_novel(novel_id, save_path)
summary = summarize_novel(save_path, ratio=summary_ratio)
```

```python
# 요약 결과 출력
print(summary)
```
