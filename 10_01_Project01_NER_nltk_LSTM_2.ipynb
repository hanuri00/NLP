```python
from nltk import word_tokenize, pos_tag, ne_chunk
sentence = 'James is working at Disney in London'
tokenized_sentence = pos_tag(word_tokenize(sentence))
print(tokenized_sentence)
```

```python
# ê°œì²´ëª… ì¸ì‹
ner_sentence = ne_chunk(tokenized_sentence)
print(ner_sentence)
```

```python
nlp = spacy.load('en_core_web_sm') # ì›í•˜ëŠ” ì–¸ì–´ì˜ ëª¨ë¸ì„ ê°€ì ¸ì˜´
doc = nlp('Apple is looking at buyin at U.K startup for $1 billion.') # ë¬¸ì¥ì„ nlp ì— ë„˜ê²¨ì£¼ë©´ ë¨
print(doc) # ì›ë˜ ë¬¸ì¥ì´ ì¶œë ¥ë¨
print(list(doc)) #ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í˜•í•˜ë©´ í† í°í™”í•œ ê²°ê³¼ê°€ ì¶œë ¥
```

```python
# NER task
doc = nlp('Apple is looking at buying U.K. startup for $1 billion')

for ent in doc.ents :
  print(ent.text, ent.label_)
```
