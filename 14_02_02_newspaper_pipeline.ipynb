```python
# 라이브러리 설치
!pip install -q transformers
!pip install -q newspaper3k

from transformers import pipeline
import newspaper

# 한국어 위키트리 URL
url = 'https://www.wikitree.co.kr/articles/897797'

# newspaper 라이브러리를 사용하여 기사 내용 추출
article = newspaper.Article(url, language='ko')
article.download()
article.parse()
content = article.text

# Transformer 모델의 pipeline 생성
summarizer = pipeline("summarization", model="t5-base", tokenizer="t5-base")

```

```python
# 문서 요약 수행
summary = summarizer(content, max_length=150, min_length=30, do_sample=False)
```

```python
# 결과 출력
print("기사 내용:\n", content[:500])  # 기사 내용의 일부분만 출력

print("\n요약:\n", summary[0]['summary_text'])
```
